# Example MATCH SSP analysis on the Milky Way GC Crater

```python
import matplotlib.pyplot as plt
%matplotlib inline
import match.scripts.asts as asts
import numpy as np
```

```python
phot = np.loadtxt('crater.phot.dat')
```

```python
plt.scatter(phot[:,0]-phot[:,1], phot[:,0], facecolor='k', edgecolor='0.8', lw=0.1, s=5)
plt.xlim(-0.5, 1.5)
plt.ylim(29,17)
plt.xlabel('F606W-F814W')
plt.ylabel('F606W')
```
